```python
text = open('./lyric.txt').read()
```

```python
import MeCab
m = MeCab.Tagger()
node = m.parse(text)
print(node)
```

```python
word_list = []
for line in node.splitlines():
    word = line.split("\t")[0]
    criteria =line.split("\t")[-1]

    if criteria.split(",")[0] in ["名詞"]:
        word_list.append(word)

print(word_list)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline

# フォントを日本語用に変更する必要がある
# mpl.matplotlib_fname()で場所を調べられる
fpath = "/Users/media/.pyenv/versions/3.9.4/lib/python3.9/site-packages/matplotlib/mpl-data/fonts/ttf/ipaexm.ttf"
stop_word_list = ["よう", "しよう", "の", "ん", "そう", "さ", "がち", "まま", "もの", "それ", "みたい"]

wc = WordCloud(background_color="white",width=900, height=500, font_path=fpath, stopwords=stop_word_list).generate(' '.join(word_list))
plt.figure(figsize=(18,10))
plt.axis("off")
plt.imshow(wc)
```
